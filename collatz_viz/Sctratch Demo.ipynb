```python
 %%manim -v WARNING --disable_caching -qm SquareToCircle

config.media_width = "75%"
config.media_embed = True

class SquareToCircle(Scene):
    def construct(self):
        circle = Circle()  # create a circle
        circle.set_fill(GREEN, opacity=0.7)  # set color and transparency
        circle.set_stroke(YELLOW, width=30)

        square = Square()  # create a square
        square.rotate(PI / 4)  # rotate a certain amount

        self.play(Create(square))  # animate the creation of the square
        self.play(Transform(square, circle))  # interpolate the square into the circle
        self.play(FadeOut(square))  # fade out animation
```

```python
 %%manim -v WARNING --disable_caching -qm FirstArcs


class FirstArcs(Scene):
    def construct(self):
        home_node = Circle()
        home_node.set_stroke(PURPLE_D, width=5)
        self.play(Create(home_node))
        self.play(home_node.animate.set_fill(PURPLE_A, opacity=1))
        target_node = Circle()
        target_node.set_stroke(BLUE_B, width=5)
        target_node.set_fill(BLUE_A, opacity=0.6)
        self.play(FadeIn(target_node))
        self.play(target_node.animate.move_to(RIGHT * 3.7))
        self.play(home_node.animate.move_to([-3, -1, 0]))
```

```python
%%manim -v WARNING --disable_caching -qm DrawTraceArc

class DrawTraceArc(MovingCameraScene):
    def construct(self):
        self.camera.frame.save_state()
        self.add(Line(5 * LEFT + 1.5 * DOWN, 5 * RIGHT + 1.5 * DOWN))
        c0 = Circle()
        c0.set_fill(GREEN_A, opacity=1.0)
        c0.set_stroke(GREEN_D, width=5)
        t0 = Text("1").next_to(c0, DOWN)
        g0 = VGroup()
        g0.add(c0).add(t0)
        self.play(FadeIn(g0))
        self.play(g0.animate.shift(3 * RIGHT))
        c1 = Circle()
        c1.set_fill(BLUE, opacity=1.0).set_stroke(BLUE_E, width=10).set_y(2)
        t1 = Text("2").next_to(c1, DOWN)
        trace1 = TracedPath(c1.get_center, stroke_width=10, stroke_color=RED)
        self.play(FadeIn(c1, t1, trace1))
        t1.add_updater(lambda x: x.next_to(c1, DOWN))
        self.play(c1.animate.shift(3 * RIGHT))
        arc_radius = norm(c1.get_center() - ORIGIN)
        arc_init_angle = np.arctan2(c1.get_center()[1], c1.get_center()[0])
        self.play(FadeIn(Line(ORIGIN, c1.get_center(), stroke_width=5, stroke_color=PURPLE),
                         Arc(radius=arc_radius, arc_center=ORIGIN, start_angle=arc_init_angle, angle=-PI/2, stroke_width=5, stroke_color=PURPLE)))
        self.wait(2)
        self.camera.add_updater(lambda x: x.auto_zoom([c0, c1, t1, trace1], margin=0.5))
        self.play(MoveAlongPath(c1, Arc(radius=arc_radius, arc_center=ORIGIN, angle=-3 * PI / 4, start_angle=arc_init_angle)))
        self.wait(4)
```
