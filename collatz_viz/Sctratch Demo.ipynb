```python
 %%manim -v WARNING --disable_caching -qm SquareToCircle

config.media_width = "75%"
config.media_embed = True

class SquareToCircle(Scene):
    def construct(self):
        circle = Circle()  # create a circle
        circle.set_fill(GREEN, opacity=0.7)  # set color and transparency
        circle.set_stroke(YELLOW, width=30)

        square = Square()  # create a square
        square.rotate(PI / 4)  # rotate a certain amount

        self.play(Create(square))  # animate the creation of the square
        self.play(Transform(square, circle))  # interpolate the square into the circle
        self.play(FadeOut(square))  # fade out animation
```

```python
 %%manim -v WARNING --disable_caching -qm FirstArcs


class FirstArcs(Scene):
    def construct(self):
        home_node = Circle()
        home_node.set_stroke(PURPLE_D, width=5)
        self.play(Create(home_node))
        self.play(home_node.animate.set_fill(PURPLE_A, opacity=1))
        target_node = Circle()
        target_node.set_stroke(BLUE_B, width=5)
        target_node.set_fill(BLUE_A, opacity=0.6)
        self.play(FadeIn(target_node))
        self.play(target_node.animate.move_to(RIGHT * 3.7))
        self.play(home_node.animate.move_to([-3, -1, 0]))
```

```python
%%manim -v WARNING --disable_caching -qm DrawTraceArc

class DrawTraceArc(MovingCameraScene):
    def construct(self):
        self.camera.frame.save_state()
        self.add(Line(5 * LEFT + 1.5 * DOWN, 5 * RIGHT + 1.5 * DOWN))
        c0 = Circle()
        c0.set_fill(GREEN_A, opacity=1.0)
        c0.set_stroke(GREEN_D, width=5)
        t0 = Text("1").next_to(c0, DOWN)
        g0 = VGroup()
        g0.add(c0).add(t0)
        self.play(FadeIn(g0))
        self.play(g0.animate.shift(3 * RIGHT))
        c1 = Circle()
        c1.set_fill(BLUE, opacity=1.0).set_stroke(BLUE_E, width=10).set_y(2)
        t1 = Text("2").next_to(c1, DOWN)
        trace1 = TracedPath(c1.get_center, stroke_width=10, stroke_color=RED)
        self.play(FadeIn(c1, t1, trace1))
        t1.add_updater(lambda x: x.next_to(c1, DOWN))
        self.play(c1.animate.shift(3 * RIGHT))
        arc_radius = norm(c1.get_center() - ORIGIN)
        arc_init_angle = np.arctan2(c1.get_center()[1], c1.get_center()[0])
        self.play(FadeIn(Line(ORIGIN, c1.get_center(), stroke_width=5, stroke_color=PURPLE),
                         Arc(radius=arc_radius, arc_center=ORIGIN, start_angle=arc_init_angle, angle=-PI/2, stroke_width=5, stroke_color=PURPLE)))
        self.wait(2)
        # self.camera.add_updater(lambda x: x.auto_zoom([c0, c1, t1, trace1], margin=0.5))
        self.play(MoveAlongPath(c1, Arc(radius=arc_radius, arc_center=ORIGIN, angle=-3 * PI / 4, start_angle=arc_init_angle)))
        self.wait(4)
```

```python
%%manim -v WARNING --disable_caching -qm DualParallelScene

class DualParallelScene(MovingCameraScene):
    def construct(self):
        self.camera.frame.save_state()
        x_axis = NumberLine([-10, 10, 1], numbers_with_elongated_ticks=range(-10, 10, 5), include_numbers=True, longer_tick_multiple=3)
        y_axis = NumberLine([-10, 10, 1], numbers_with_elongated_ticks=range(-10, 10, 5), include_numbers=True, longer_tick_multiple=3, rotation=90 * DEGREES, label_direction=LEFT)
        self.add(x_axis, y_axis)
        d0 = Dot(color=GREEN, fill_opacity=0.7, radius=0.2)
        d1 = Dot(color=BLUE, fill_opacity=0.7, radius=0.2)
        self.play(FadeIn(d0, d1))
        self.play(AnimationGroup(d0.animate.move_to([0, 3, 0]), d1.animate.move_to([3, 0, 0])))
        
```

```python
 %%manim -v WARNING --disable_caching -qm DotDebug
    
config.media_width = "75%"
config.media_embed = True

class DotDebug(MovingCameraScene):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.circle_radius = 1.0
        self.trace_stroke_width = 4
        self.circle_fill_color_start = DARK_BLUE
        self.circle_fill_color_end = RED
        # How opaque to color each node circle.
        self.circle_fill_opacity = 1.0
        # Color of node edge.
        self.circle_stroke_color = PURPLE
        # Thickness of node edge.
        self.circle_stroke_width = 8
        # How big of location dots to render on the number line.
        self.dot_radius = 0.45
        # Dot opacity.
        self.dot_opacity = 0.7
        # Color of trace that we draw as we move nodes.
        self.trace_stroke_color = BLUE_A
        # Thickness of trace.
        self.trace_stroke_width = 8
        self.circle_radius_decay = 0.95

    def get_color_by_shell(self, shell: int) -> mnm.color.Color:
        """Generate a color scaled by radial shell from the origin.

        :param shell: The tree depth (i.e., shell) at which we're rendering.
        :type shell: int
        :return: Color for this shell, interpolated between start and end colors.
        :rtype: mnm.color.Color
        """
        decay = np.power(self.circle_radius_decay, shell)
        return utils.color.interpolate_color(self.circle_fill_color_start,
                                                 self.circle_fill_color_end,
                                                 1.0 - decay)

    def construct(self):
        shell = 0
        nl = NumberLine(x_range=[0, 64, 2],
                                numbers_with_elongated_ticks=range(0, 64, 2),
                                longer_tick_multiple=3,
                                include_tip=True,
                                tip_shape=ArrowCircleTip,
                                include_numbers=True,
                                stroke_width=2 * self.trace_stroke_width,
                                font_size=64,
                                line_to_number_buff=LARGE_BUFF)
        nl.move_to(DOWN * 2.5 * self.circle_radius - nl.number_to_point(1) + IN)
        self.add(nl)
        c = Circle(radius=self.circle_radius)
        c.set_stroke(color=self.circle_fill_color_start, width=self.circle_stroke_width)
        c.set_fill(color=self.get_color_by_shell(shell=shell), opacity=self.circle_fill_opacity)
        d = Dot(radius=self.dot_radius,
                       fill_opacity=self.dot_opacity,
                       color=self.get_color_by_shell(shell=shell))
        d.move_to(nl.number_to_point(5))
        self.play(FadeIn(c, d))
        self.play(d.animate.move_to(nl.number_to_point(1.2)))
        self.wait(1.5)

```
